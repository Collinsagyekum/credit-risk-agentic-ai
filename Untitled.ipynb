```python
import pandas as pd
import numpy as np
import joblib

# Load the artifacts we saved on Day 2
model = joblib.load("models/credit_risk_model.pkl")
feature_columns = joblib.load("models/feature_columns.pkl")

print("Model loaded:", type(model).__name__)
print("Expected feature count:", len(feature_columns))
print("First few features:", feature_columns[:5])
```

```python
# A sample raw applicant вЂ” the kind of input the agent will eventually send
sample_applicant = {
    "checking_account": "A11",      # < 0 DM (weak)
    "duration_months": 24,
    "credit_history": "A32",        # existing credits paid back duly
    "purpose": "A43",               # radio/television
    "credit_amount": 5000,
    "savings_account": "A61",       # < 100 DM
    "employment_since": "A73",      # 1-4 years
    "installment_rate": 4,
    "personal_status_sex": "A93",   # male, single
    "other_debtors": "A101",        # none
    "residence_since": 2,
    "property": "A121",             # real estate
    "age": 35,
    "other_installment_plans": "A143",  # none
    "housing": "A152",              # own
    "existing_credits": 1,
    "job": "A173",                  # skilled employee
    "num_dependents": 1,
    "telephone": "A192",            # yes
    "foreign_worker": "A201",       # yes
}

def encode_applicant(applicant: dict, feature_columns: list) -> pd.DataFrame:
    """Turn a raw applicant dict into the exact model-ready feature row."""
    # One row dataframe from the raw input
    raw = pd.DataFrame([applicant])
    # One-hot encode the same way as training
    encoded = pd.get_dummies(raw)
    # Align to the trained columns: add missing as 0, drop extras, fix order
    aligned = encoded.reindex(columns=feature_columns, fill_value=0)
    return aligned

# Test it
encoded_row = encode_applicant(sample_applicant, feature_columns)
print("Encoded shape:", encoded_row.shape, "(should be 1 row,", len(feature_columns), "columns)")
print("Any missing columns?", encoded_row.isnull().sum().sum(), "(should be 0)")
```

```python
def score_applicant(applicant: dict) -> dict:
    """Score a loan applicant. Returns risk probability, decision, and top factors."""
    # Encode and align
    X = encode_applicant(applicant, feature_columns)

    # Predict probability of default (class 1)
    prob_default = float(model.predict_proba(X)[0, 1])

    # Decision based on a threshold (a business lever, not a fixed truth)
    threshold = 0.5
    decision = "DECLINE" if prob_default >= threshold else "APPROVE"

    # Risk tier for human readability
    if prob_default < 0.3:
        tier = "Low"
    elif prob_default < 0.6:
        tier = "Medium"
    else:
        tier = "High"

    # Top contributing factors: which of THIS applicant's active features
    # carry the most model importance
    importances = dict(zip(feature_columns, model.feature_importances_))
    active = X.iloc[0]
    active_features = [f for f in feature_columns if active[f] != 0]
    top_factors = sorted(active_features, key=lambda f: importances[f], reverse=True)[:5]

    return {
        "probability_of_default": round(prob_default, 3),
        "risk_tier": tier,
        "decision": decision,
        "threshold_used": threshold,
        "top_factors": top_factors,
    }

# Test it on our sample applicant
import json
result = score_applicant(sample_applicant)
print(json.dumps(result, indent=2))
```

```python
import shap

# Build the explainer once (TreeExplainer is fast and exact for XGBoost)
explainer = shap.TreeExplainer(model)

def score_applicant(applicant: dict) -> dict:
    """Score an applicant with SHAP-based per-prediction explanations."""
    X = encode_applicant(applicant, feature_columns)
    prob_default = float(model.predict_proba(X)[0, 1])

    threshold = 0.5
    decision = "DECLINE" if prob_default >= threshold else "APPROVE"
    if prob_default < 0.3:
        tier = "Low"
    elif prob_default < 0.6:
        tier = "Medium"
    else:
        tier = "High"

    # --- SHAP per-prediction attribution ---
    try:
        shap_values = explainer.shap_values(X)[0]   # one row
        contribs = sorted(
            zip(feature_columns, shap_values),
            key=lambda x: abs(x[1]), reverse=True
        )
        # Signed: positive pushes toward default, negative toward approval
        increased_risk = [f for f, v in contribs if v > 0][:3]
        decreased_risk = [f for f, v in contribs if v < 0][:3]
        method = "shap"
    except Exception as e:
        # Fallback: global-importance among active features
        importances = dict(zip(feature_columns, model.feature_importances_))
        active = X.iloc[0]
        active_feats = [f for f in feature_columns if active[f] != 0]
        increased_risk = sorted(active_feats, key=lambda f: importances[f], reverse=True)[:3]
        decreased_risk = []
        method = f"fallback ({type(e).__name__})"

    return {
        "probability_of_default": round(prob_default, 3),
        "risk_tier": tier,
        "decision": decision,
        "threshold_used": threshold,
        "factors_increasing_risk": increased_risk,
        "factors_decreasing_risk": decreased_risk,
        "explanation_method": method,
    }

# Test on the same applicant
import json
result = score_applicant(sample_applicant)
print(json.dumps(result, indent=2))
```
